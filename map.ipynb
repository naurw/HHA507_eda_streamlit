# Creating an simple yet interactive map to visualize the locations of the hospitals within the csv file that will be analyzed within this repository 

```python
import pandas as pd
import folium
```

```python
hospital = pd.read_csv('https://raw.githubusercontent.com/hantswilliams/AHI_STATS_507/main/Week13_Summary/output/df_hospital_2.csv')
```

## Isolating the state of interest

```python
nyHospitals = hospital[hospital['state'] == 'NY']
```

```python
nyHospitals[['Point', 'Longitude', 'Latitude']] = nyHospitals['location'].str.strip('()').str.split(' ', expand=True).rename(columns={0: 'Point', 1:'Longitude', 2:'Latitude'}) 
nyHospitals['Longitude'] = nyHospitals['Longitude'].str.strip('(')
nyHospitals['Longitude'] = nyHospitals['Longitude'].astype(float)
nyHospitals['Latitude'] = pd.to_numeric(nyHospitals['Latitude'])
```

```python
nyHospitals.drop(['location'], axis =1, inplace=True)
```

```python
nyHospitals.info()
```

```python
hMap = nyHospitals[['hospital_name', 'Longitude', 'Latitude']]
hMap = hMap.dropna()
```

```python
hMap
```

```python
map = folium.Map(location = [hMap.Latitude.mean(), hMap.Longitude.mean()], zoom_start= 14, control_scale= True)
for index, location_info in hMap.iterrows():
    folium.Marker([location_info['Latitude'], location_info['Longitude']], popup= location_info['hospital_name']).add_to(map)
```
